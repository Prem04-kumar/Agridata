```python

def detect_columns(crop):
    """Detect area and production columns for a given crop."""
    area_cols = [c for c in df.columns if crop.lower() in c.lower() and "area" in c.lower()]
    prod_cols = [c for c in df.columns if crop.lower() in c.lower() and "production" in c.lower()]
    if not area_cols:
        raise ValueError(f"No {crop} Area column found. Please check your CSV.")
    if not prod_cols:
        raise ValueError(f"No {crop} Production column found. Please check your CSV.")
    return area_cols[0], prod_cols[0]

rice_area_col, rice_prod_col   = detect_columns("rice")
wheat_area_col, wheat_prod_col = detect_columns("wheat")
maize_area_col, maize_prod_col = detect_columns("maize")

print("Using columns:")
print("Rice Area:", rice_area_col, "| Rice Production:", rice_prod_col)
print("Wheat Area:", wheat_area_col, "| Wheat Production:", wheat_prod_col)
print("Maize Area:", maize_area_col, "| Maize Production:", maize_prod_col)
```

```python
def detect_columns(crop):
    area_cols = [c for c in df.columns if crop.lower() in c.lower() and "area" in c.lower()]
    prod_cols = [c for c in df.columns if crop.lower() in c.lower() and "production" in c.lower()]
    if not area_cols:
        raise ValueError(f"No {crop} Area column found. Please check your CSV.")
    if not prod_cols:
        raise ValueError(f"No {crop} Production column found. Please check your CSV.")
    return area_cols[0], prod_cols[0]

rice_area_col, rice_prod_col   = detect_columns("rice")
wheat_area_col, wheat_prod_col = detect_columns("wheat")


print("Using columns:")
print("Rice Area:", rice_area_col, "| Rice Production:", rice_prod_col)
print("Wheat Area:", wheat_area_col, "| Wheat Production:", wheat_prod_col)

```
